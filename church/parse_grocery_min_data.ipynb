```python
import pandas as pd
import re
```

```python
data = pd.read_excel("10_17 Grocery Sign-in updated.xlsx", keep_default_na=False)
display(data)
```

```python
print(data.columns)
zero_17 = []
eighteen_64 = []
sixtyfive_plus = []

def extract_count(raw: str) -> int:
    count = 0

    if raw:
        c = re.search(r'_+(\d)', raw)
        if c and len(c.groups()) > 0:
            count = c.groups()[0]

    print(f"'{raw.strip()}' --> {count}")
    return count

col_name = "# of Family Members in age groups -                       miembros de la familia"
for raw_col_value in data[col_name]:
    print('------------------------------------')
    # print(raw_col_value)
    # print([x for x in raw_col_value])
    regex = ".*?0-17_*?(\d)_*?|_*?.*?18-64.*?_*?(\d)_*?|_*?.*?65\+.*?_*?(\d)_*?|_*?"
    if raw_col_value:
        raw_col_vals = raw_col_value.split('#')
        regex = r"-.*?_+(\d)_*$"
        zero_17.append(extract_count(raw_col_vals[1]))
        eighteen_64.append(extract_count(raw_col_vals[2]))
        sixtyfive_plus.append(extract_count(raw_col_vals[3]))
    else:
        zero_17.append(0)
        eighteen_64.append(0)
        sixtyfive_plus.append(0)
        
            
# counts = [re.search(regex, raw).groups() for raw in raw_col_vals]
```

```python
data['Age 0-17'] = pd.Series(zero_17)
data['Age 18-64'] = pd.Series(eighteen_64)
data['Age 65+'] = pd.Series(sixtyfive_plus)
```

```python
display(data)
data.to_excel("UpdatedData.xlsx")
```
